```python
df.drop(columns = ['PassengerId','Name','Ticket','Fare','Cabin'],inplace = True)
```

```python
df['Survived'].fillna(method = 'ffill',inplace = True)
```

```python
df.isnull().sum()
```

```python
x = pd.DataFrame(df.iloc[:,:])
x
```

```python
x.drop(columns = ['Survived'],inplace = True)
x
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
tr = ColumnTransformer(transformers=[
    ('t1',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked'])
],remainder = 'passthrough')
```

```python
x1 = tr.fit_transform(x)
x1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
a = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,a))
```

```python
pd.DataFrame(confusion_matrix(y_test,a))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Decision Tree Prediction'] = a
result
```
