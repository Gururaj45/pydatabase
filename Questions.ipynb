```python
import pandas as pd
from pandas import DataFrame
df = DataFrame([
                ['Aryan','singh','Full-time','adminstration',2,20000],
                ['Rohan','Agarwal','intern','Technical',3,5000],
                ['Riya','shah','Full-time','adminstration',5,10000 ],
                ['Yash','Bhatia','part-time','Technical',7,10000 ],
                ['Siddant','Khanna','Full-time','Management',6,20000]
                             ]
                               ,columns=['First Name','Last Name','Type','Department','YoE','Salary']   
                            )
df
```

1. Make a pivot table which shows average salary of each type of employee for each department

```python
import pandas as pd
table = pd.pivot_table(df,index="Type",columns="Department",values="Salary",aggfunc="mean")
table
```

2. Make a pivot table which shows the sum and mean of the salaries of each type of 
employee and the number of employees of each type.

```python
table1 = pd.pivot_table(df,index="Type",values='Salary',aggfunc=['sum','mean','count'])
table1
```
