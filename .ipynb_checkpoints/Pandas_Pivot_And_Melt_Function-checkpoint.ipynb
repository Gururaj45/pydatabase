```python
import pandas as pd
from pandas import DataFrame
df1 = DataFrame([
                ['Aryan','singh','Full-time','adminstration',2,20000],
                ['Rohan','Agarwal','intern','Technical',3,5000],
                ['Riya','shah','Full-time','adminstration',5,10000 ],
                ['Yash','Bhatia','part-time','Technical',7,10000 ],
                ['Siddant','Khanna','Full-time','Management',6,20000]
                             ]
                               ,columns=['First Name','Last Name','Type','Department','YoE','Salary']   
                            )
df1
```

```python
table = pd.pivot_table(df1,index="Type")
table
```

```python
table = pd.pivot_table(df1,index="Type",values="Salary")
table
```

```python
table = pd.pivot_table(df1,index="Type",columns="Department")
table
```

```python
table = pd.pivot_table(df1,index="Type",columns='Department',values='Salary')
table
```

```python
table = pd.pivot_table(df1,index="Type",columns='Department',values='Salary',aggfunc='sum')
table
```
